```python
import json
from pathlib import Path
import sys
from datetime import datetime
```

```python
sys.path.append('../')
```

```python
from generate_from_notebook import generate_from_notebook
```

```python
def read_json(path: Path) -> dict:
    with path.open("r", encoding="utf-8") as f:
        return json.load(f)
```

```python
combined_art_styles = read_json(Path('../samples/combined_art_styles.json'))
```

```python
model_name="sd35_base"

for category in combined_art_styles:
    print(f"### {category} ###")
    for style in combined_art_styles[category]:
        print(f"## {style} ##")
        style_prompt = combined_art_styles[category][style]['style prompt']
        for example in combined_art_styles[category][style]['examples']:
            example = example.replace('A ', '').replace('a ', '')

            prompt = style_prompt + ', ' + example
            print(f"# {prompt} #")

            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

            image_path = f"../samples/test/{style.replace('/', '_').replace(' ', '_')}_{model_name}_{timestamp}.png"
            print(image_path)

            result = generate_from_notebook(
                prompt=prompt,
                model_name=model_name,
                output_file=image_path,
                width=1024,
                height=1024,
                steps=40,
                guidance_scale=4,
                seed=None,
                negative_prompt="blurry, low quality, cluttered background",
                image_count=1,
                overwrite=False,
            )
            print(f"image saved at {result["image_paths"][0]}")
```
